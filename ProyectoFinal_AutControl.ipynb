```python
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import pandas as pd

# Supongamos que ya tienes un DataFrame llamado 'datos_csv_enbruto_clon'
# Asegúrate de que este DataFrame contenga las columnas 'Vrms' y 'Clasificacion'

dataset = pd.DataFrame(datos_csv_enbruto_clon)

# Dividir el conjunto de datos en características (X) y etiquetas (Y)
X = dataset[['Vrms']]
Y = dataset['Clasificacion']

# Dividir los datos en conjuntos de entrenamiento y prueba
X_Train, X_Test, Y_Train, Y_Test = train_test_split(X, Y, test_size=0.3, random_state=10)

# Normalizar datos
scaler = StandardScaler()
X_Train_scaled = scaler.fit_transform(X_Train)
X_Test_scaled = scaler.transform(X_Test)

# Manejo de desequilibrio de clases con SMOTE
smote = SMOTE(random_state=1)
X_Train_resampled, Y_Train_resampled = smote.fit_resample(X_Train_scaled, Y_Train)

# Asegúrate de que X_Train_resampled y X_Test_scaled tengan la forma adecuada
X_Train_resampled = X_Train_resampled.reshape(-1, 1)
X_Test_scaled = X_Test_scaled.reshape(-1, 1)

# Modelo de regresión logística
modelo_logistico = LogisticRegression(random_state=1)
modelo_logistico.fit(X_Train_resampled, Y_Train_resampled)

# Predicciones y evaluación del modelo
predicciones_logistico = modelo_logistico.predict(X_Test_scaled)

reporte_logistico = classification_report(Y_Test, predicciones_logistico)
print(predicciones_logistico)
```
